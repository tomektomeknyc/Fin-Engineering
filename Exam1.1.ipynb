Compute Drawdown
-Wealth Index
-Previous Peaks
-Drawdown as a percentage of the previous peak

```python
wealth_index = 1000*(1+ returns["Hi 20"]).cumprod()
```

```python
wealth_index.head()
```

```python
wealth_index.plot.line()
```

```python
previous_peaks = wealth_index.cummax()
```

```python
previous_peaks.plot()
```

```python
drawdown= (wealth_index - previous_peaks)/previous_peaks
drawdown.plot()
```

```python
drawdown.min()
```

```python
drawdown.max()
```
